Close to close estimator 

```python
def close_to_close(data):
    data['Close-to-close'] = np.abs(np.log(data['Close'] / data['Close'].shift(1)))
    Close_to_close=data['Close-to-close']
    return Close_to_close
```

Parkinson estimator 

```python
def Parkinson(data):
    Parkinson = np.sqrt(1 / (4 * np.log(2)) * np.log(data['High'] / data['Low']) ** 2)
    return Parkinson
```

Garman Klass estimator

```python
def garman_klass__estimator(data):
    sigma_GK = 0.5 * (np.log(data['High']/data['Low']))**2 - (2*np.log(2)-1) * (np.log(data['Close']/data['Open']))**2
    sigma_GK = np.sqrt(sigma_GK)
    return sigma_GK
```

Roger Satchell estimator

```python
def rogers_satchell_volatility(data):
    high = data['High']
    low = data['Low']
    close = data['Close']
    open = data['Open']

    log_hc = np.log(high / close)
    log_ho = np.log(high / open)
    log_lc = np.log(low / close)
    log_lo = np.log(low / open)

    sigma_rs = np.sqrt(log_hc * log_ho + log_lc * log_lo)
    return sigma_rs
```

Yang-Zhang estimator

```python

def calculate_yang_zhang(data, k):
    yang_zhang = np.sqrt(
        np.log(data['Open'] / data['Close'].shift(1)) ** 2
        + k * np.log(data['Close'] / data['Open']) ** 2
        + (1 - k) * (np.log(data['High'] / data['Close']) * np.log(data['High'] / data['Open'])
                  + np.log(data['Low'] / data['Close']) * np.log(data['Low'] / data['Open'])))
    
    return yang_zhang

# Variance of the Yang-Zhang indicator as a function of k
def plot_yang_zhang_variance(data):
    k_values = np.linspace(0, 1, 100)
    variances = []

    for k in k_values:
        yang_zhang = calculate_yang_zhang(data, k)
        variances.append(np.var(yang_zhang))

    plt.figure(figsize=(10, 6))
    plt.plot(k_values, variances, label='Yang-Zhang variance')
    plt.xlabel('k')
    plt.ylabel('Variance')
    plt.title('Variance of the Yang-Zhang indicator as a function of k')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_yang_zhang_variance(data)
```

III) Garch model

Garch prediction

```python

data = pd.read_csv('RealizedVarianceData.csv', low_memory=False)


data['Unnamed: 0'] = pd.to_datetime(data['Unnamed: 0'], errors='coerce')
data = data.dropna(subset=['Unnamed: 0', 'Realized Variance (5-minute)'])


data['Realized Variance (5-minute)'] = pd.to_numeric(data['Realized Variance (5-minute)'], errors='coerce')
data = data.dropna(subset=['Realized Variance (5-minute)'])


data.rename(columns={'Unnamed: 0': 'Date'}, inplace=True)
data.set_index('Date', inplace=True)


data['Realized Volatility (5-minute)'] = np.sqrt(data['Realized Variance (5-minute)']*100)


split_date = '2010-01-01'
in_sample = data.loc[:split_date]
out_sample = data.loc[split_date:]

# Parameters
a=2
b=0
# Rolling window forecast
window_size = 1
forecasts = []

for start in range(len(out_sample) - window_size):
    end = start + window_size
    train_data = pd.concat([in_sample['Realized Volatility (5-minute)'], out_sample['Realized Volatility (5-minute)'][:start]])
    
    # Fit the GARCH model
    model = arch_model(train_data, vol='Garch', p=a, q=b)
    model_fit = model.fit(disp='off')
    if b==0:    
       model_fit.params['beta[1]']=0
    last_volatility = train_data.iloc[-1]
    forecast_variance = model_fit.params['omega'] + model_fit.params['alpha[1]'] * (last_volatility ** 2) + model_fit.params['beta[1]'] * model_fit.conditional_volatility[-1] ** 2
    forecast_volatility = np.sqrt(forecast_variance)
    forecasts.append(forecast_volatility)

forecasts = pd.Series(forecasts, index=out_sample.index[window_size:])
forecasts=forecasts.shift(-2)

plt.figure(figsize=(12, 6))
plt.plot(forecasts.index, out_sample['Realized Volatility (5-minute)'][window_size:], label='True Out-Sample Volatility')
plt.plot(forecasts.index, forecasts, label='Forecasted Volatility')
plt.legend()
plt.title('True vs Forecasted Volatility')
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.show()
```

Garch study

```python
warnings.filterwarnings("ignore")

def download_data(ticker='SPY', start='2000-01-01', end='2013-12-31'):
    data = yf.download(ticker, start=start, end=end)
    data = data[['Open', 'High', 'Low', 'Close']]
    return data

# GARCH
def apply_garch_model(log_returns, p, q):
    p, q = int(p), int(q)
    model = arch_model(log_returns, vol='Garch', p=p, q=q, rescale=False)
    model_fit = model.fit(disp="off")
    forecasts = model_fit.forecast(horizon=1, start=0)
    predicted_vol = np.sqrt(forecasts.variance.values.flatten())
    return predicted_vol

# ARCH
def apply_arch_model(log_returns, p):
    p = int(p)
    model = arch_model(log_returns, vol='ARCH', p=p, rescale=False)
    model_fit = model.fit(disp="off")
    forecasts = model_fit.forecast(horizon=1, start=0)
    predicted_vol = np.sqrt(forecasts.variance.values.flatten())
    return predicted_vol

# MSE
def calculate_weekly_mse(realized_vol, predicted_vol):
    mse_weekly = ((realized_vol - predicted_vol) ** 2).resample('W').mean()
    return mse_weekly


data = download_data()
    
log_returns = np.log(data['Close'] / data['Close'].shift(1)).dropna()
    
realized_vol = log_returns.abs()

results = []

# GARCH for p, q in [1, 2, 3]
for p in range(1, 4):
    for q in range(1, 4):
        predicted_vol_garch = apply_garch_model(log_returns, p, q)
        predicted_vol_garch = predicted_vol_garch[-len(realized_vol):]
        mse_weekly_garch = calculate_weekly_mse(realized_vol, predicted_vol_garch)
        avg_mse_garch = mse_weekly_garch.mean()
        results.append(('GARCH', p, q, avg_mse_garch))

# ARCH for p in [1, 2, 3]
for p in range(1, 4):
    predicted_vol_arch = apply_arch_model(log_returns, p)
    predicted_vol_arch = predicted_vol_arch[-len(realized_vol):]
    mse_weekly_arch = calculate_weekly_mse(realized_vol, predicted_vol_arch)
    avg_mse_arch = mse_weekly_arch.mean()
    results.append(('ARCH', p, 0, avg_mse_arch)) 

results_df = pd.DataFrame(results, columns=['Model', 'p', 'q', 'Avg_MSE'])
    
print(results_df)

num_models = len(results_df)
num_cols = 4
num_rows = (num_models + num_cols - 1) // num_cols 

fig, axs = plt.subplots(num_rows, num_cols, figsize=(18, num_rows * 6))
fig.suptitle('Histograms of weekly MSE for GARCH and ARCH models')

for index, (model, p, q, avg_mse) in enumerate(results):
    if model == 'GARCH':
        predicted_vol = apply_garch_model(log_returns, p, q)
    else:
        predicted_vol = apply_arch_model(log_returns, p)
        
    predicted_vol = predicted_vol[-len(realized_vol):]
    mse_weekly = calculate_weekly_mse(realized_vol, predicted_vol)
        
    ax = axs[index // num_cols, index % num_cols]
    ax.hist(mse_weekly.dropna(), bins=50, edgecolor='black')
    ax.set_title(f'{model} (p={p}, q={q})')
    ax.set_xlabel('MSE')
    ax.set_ylabel('Frequency')
    ax.text(0.95, 0.95, f'Avg MSE: {avg_mse:.6f}', horizontalalignment='right', verticalalignment='top', transform=ax.transAxes, fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

for i in range(num_models, num_rows * num_cols):
    fig.delaxes(axs.flatten()[i])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Arch and garch bias

```python
# weekly variance
def calculate_weekly_variance(volatility, index):
    volatility_series = pd.Series(volatility, index=index)
    var_weekly = volatility_series.resample('W').var()
    return var_weekly

# weekly bias
def calculate_weekly_bias(realized_vol, predicted_vol, index):
    var_realized_weekly = calculate_weekly_variance(realized_vol, index)
    var_predicted_weekly = calculate_weekly_variance(predicted_vol, index)
    bias_weekly = var_realized_weekly - var_predicted_weekly
    return bias_weekly


data = download_data()
    
log_returns = np.log(data['Close'] / data['Close'].shift(1)).dropna()
    
realized_vol = log_returns.abs()
  
results = []

# GARCH for p, q in [1, 2, 3]
for p in range(1, 4):
    for q in range(1, 4):
        predicted_vol_garch = apply_garch_model(log_returns, p, q)
        predicted_vol_garch = predicted_vol_garch[-len(realized_vol):]
        bias_weekly_garch = calculate_weekly_bias(realized_vol, predicted_vol_garch, realized_vol.index)
        avg_bias_garch = bias_weekly_garch.mean()
        results.append(('GARCH', p, q, avg_bias_garch))

# ARCH for p in [1, 2, 3]
for p in range(1, 4):
    predicted_vol_arch = apply_arch_model(log_returns, p)
    predicted_vol_arch = predicted_vol_arch[-len(realized_vol):]
    bias_weekly_arch = calculate_weekly_bias(realized_vol, predicted_vol_arch, realized_vol.index)
    avg_bias_arch = bias_weekly_arch.mean()
    results.append(('ARCH', p, 0, avg_bias_arch))

results_df = pd.DataFrame(results, columns=['Model', 'p', 'q', 'Avg_Weekly_Bias'])
    
print(results_df)

num_models = len(results_df)
num_cols = 4
num_rows = (num_models + num_cols - 1) // num_cols 

fig, axs = plt.subplots(num_rows, num_cols, figsize=(18, num_rows * 6))
fig.suptitle('Histograms of weekly biases for GARCH and ARCH models')

for index, (model, p, q, avg_bias) in enumerate(results):
    if model == 'GARCH':
        predicted_vol = apply_garch_model(log_returns, p, q)
    else:
        predicted_vol = apply_arch_model(log_returns, p)
        
    predicted_vol = predicted_vol[-len(realized_vol):]
    bias_weekly = calculate_weekly_bias(realized_vol, predicted_vol, realized_vol.index)
        
    ax = axs[index // num_cols, index % num_cols]
    ax.hist(bias_weekly.dropna(), bins=50, edgecolor='black')
    ax.set_title(f'{model} (p={p}, q={q})')
    ax.set_xlabel('Bias')
    ax.set_ylabel('Frequency')
    ax.text(0.95, 0.95, f'Avg Bias: {avg_bias:.6f}', horizontalalignment='right', verticalalignment='top', transform=ax.transAxes, fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

for i in range(num_models, num_rows * num_cols):
    fig.delaxes(axs.flatten()[i])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
def M(X,q,lag):
    a=np.mean(abs(np.diff(X[::int(lag)]))**q)
    return(a)

def H_exp_model(X_sim,plot):
    Delta=np.arange(1,251,1)
    Q=[0.5,1,1.5,2,3]
    Y=np.zeros(len(Delta)) 
    h_exp=0
    for q in Q:
        i=0
        for d in Delta:
            Em=0
            for X in X_sim:
                Em+=M(X,q,d)
            Y[i]=Em/len(X_sim)
            i+=1
        Y=np.log10(Y)
        slope, intercept,a,b,c= stats.linregress(np.log10(Delta),Y)
        x=[0,2.5]
        y=[0*slope+intercept,2.5*slope+intercept]
        plt.plot(x,y)
        h_exp+=slope/q/len(Q)
        plt.scatter(np.log10(Delta),Y,marker=".",s=10)
    plt.title("H theoric = "+str(H)+", H experimental = "+str(h_exp))
    plt.xlabel(r"$\log(\Delta)$")
    plt.ylabel(r"$\log(m(q,\Delta))$")
    if plot:
        plt.show()
    else:
        plt.clf()
    return(h_exp)
```

```python
def autocorr_empirique(x,delta):
    mean = np.mean(x)
    n=len(x)
    z=(x[:n-delta]-mean)*(x[delta:]-mean)
    return(np.mean(z))

def autocorr(x_simu,delta):
    mea=0
    for x in x_simu:
        mea+=autocorr_empirique(x,delta)
    return(mea/len(x_simu))

a=X_simu[:,:8000]
a=a**10
Delta=np.arange(1000)
Delta2=np.arange(len(D)-2600)

A1=[autocorr(a,delta) for delta in Delta]

B1=[np.log10(autocorr(a,delta)) for delta in Delta]

C1=[autocorr(np.log10(a),delta) for delta in Delta]

A2=[autocorr_empirique(D,delta) for delta in Delta2]

B2=[np.log10(autocorr_empirique(D,delta)) for delta in Delta2]

C2=[autocorr_empirique(np.log10(D),delta) for delta in Delta2]
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

def simulate_gains(alpha, prices):
    stocks = 1 
    cash = 0
    total_values = []
    for k in range(1, len(prices)):
        total_values.append(stocks * prices[k] + cash)
        if prices[k] - prices[k-1] > alpha and stocks > 0:  # Vendre
            cash = stocks * prices[k]  # Vendre tout
            stocks = 0
        elif prices[k] - prices[k-1] < -alpha and cash > 0:  # Acheter
            stocks = cash / prices[k]
            cash = 0
    return total_values


# Fonction pour simuler l'volution en fonction de la fentre de roulement
def simulate_evolution(prices, window_size):
    sell_indices, sell_prices, buy_indices, buy_prices = [], [], [], []
    stocks = 1 
    cash = 0
    total_values = []
    alpha = calculate_optimal_alpha(prices[:window_size])
    for k in range(window_size, len(prices)):
        alpha = calculate_optimal_alpha(prices[k-window_size:k])
        total_values.append(stocks * prices[k] + cash)
        if prices[k] - prices[k-1] > alpha and stocks > 0:  # Vendre
            sell_indices.append(k)
            sell_prices.append(prices[k])
            cash = stocks * prices[k]  # Vendre tout
            stocks = 0
        elif prices[k] - prices[k-1] < -alpha and cash > 0:  # Acheter
            buy_indices.append(k)
            buy_prices.append(prices[k])
            stocks = cash / prices[k]
            cash = 0
    return total_values, sell_indices, sell_prices, buy_indices, buy_prices

# Fonction pour calculer l'alpha optimal
def calculate_optimal_alpha(prices):
    max_profit = 0
    optimal_alpha = 0
    for k in np.arange(-0.1, 0.1, 0.01):  # Ajuster l'intervalle et le pas si ncessaire
        total_values = simulate_gains(k, prices)
        if total_values[-1] > max_profit:
            optimal_alpha = k
            max_profit = total_values[-1]
    return optimal_alpha

# Fonction pour tester la stratgie sur un actif
def test_alpha_strategy(ticker, start_date, end_date, rolling_window=10):
    data = yf.download(ticker, start=start_date, end=end_date)
    if data.empty:
        print(f"Aucune donne pour {ticker}")
        return None

    prices = data['Close'].values
    total_values, sell_indices, sell_prices, buy_indices, buy_prices = simulate_evolution(prices, rolling_window)
    total_values_with_init = np.array([0 for _ in range(rolling_window)] + total_values)

    # Calculer les rendements du portefeuille
    portfolio_returns = np.diff(total_values_with_init) / total_values_with_init[:-1]
   
    # Calculer le Sharpe Ratio
 
    sharpe_ratio = np.mean(portfolio_returns) / np.std(portfolio_returns)

    # Calculer le drawdown
    cumulative_returns = np.cumprod(1 + portfolio_returns)
    peak = np.maximum.accumulate(cumulative_returns)
    drawdown = (cumulative_returns - peak) / peak
    max_drawdown = np.max(drawdown)

    # Tracer l'volution du portefeuille et des prix avec les signaux d'achat et de vente
    plt.figure(figsize=(14, 7))
    plt.plot(prices, color='darkgrey', label="Prix de l'actif")
    plt.plot(total_values_with_init, color='grey', label="Nombre d'actions * Prix d'une action + Liquidits")
    plt.scatter(sell_indices, sell_prices, color='red', marker='v', label='Vendre ' + str(len(sell_indices)) + ' fois')
    plt.scatter(buy_indices, buy_prices, color='green', marker='^', label='Acheter ' + str(len(buy_indices)) + ' fois')
    plt.title(f"Gain d'argent d'une action avec une stratgie Alpha avec une fentre de {rolling_window} jours pour {ticker}")
    plt.legend()
    plt.xlabel('Temps')
    plt.ylabel('Valeur')
    plt.grid(True)
    plt.show()

    # Calculer le profit total
    buy_prices = [buy_prices[i] for i in range(len(buy_prices))]
    sell_prices = [sell_prices[i] for i in range(len(sell_prices))]
    num_trades = min(len(buy_prices), len(sell_prices))
    profits = np.array(sell_prices[:num_trades]) - np.array(buy_prices[:num_trades])
    total_profit = profits.sum()

    # Retourner les rsultats sous forme de dictionnaire
    results = {
        'Profit total': total_profit,
        'Nombre de transactions': num_trades,
    }

    return results

```

```python
# Tester la stratgie sur le FTSE  pour la priode entre 2010 et 2013
FTSE = test_alpha_strategy('^FTSE', '2000-01-01', '2013-12-31')
print("Rsultats pour le NASDAQ:", FTSE)

# Tester la stratgie sur le SPY (ETF S&P 500) pour la priode entre 2010 et 2013
spy_results = test_alpha_strategy('SPY', '2000-01-01', '2013-12-31')
print("Rsultats pour le SPY:", spy_results)
```

RSI strategy 

We will evaluate the strategy thanks to some indicators : cumulative PandL, Sharp Ratio and the drawdown

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def analyze_trading_strategy_RSI(symbol, start_date='2020-01-01', end_date='2024-06-01', rsi_period=14, rsi_overbought=70, rsi_oversold=30):
    # Tlchargement des donnes historiques
    historical_data = yf.download(symbol, start=start_date, end=end_date)
    print("Historical data downloaded for", symbol)
    
    # Calcul de l'indicateur RSI
    delta = historical_data['Adj Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=rsi_period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=rsi_period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    historical_data['RSI'] = rsi

    # Gnration des signaux d'achat et de vente bass sur RSI
    historical_data['Signal'] = np.where(historical_data['RSI'] < rsi_oversold, 1, np.where(historical_data['RSI'] > rsi_overbought, -1, 0))
    historical_data['Signal'] = historical_data['Signal'].diff().fillna(0)
    historical_data['Buy_Price'] = np.where((historical_data['Signal'] == 1), historical_data['Adj Close'], np.nan)
    historical_data['Sell_Price'] = np.where((historical_data['Signal'] == -1), historical_data['Adj Close'], np.nan)

    # Calcul du P&L quotidien et cumulatif
    historical_data['Return'] = historical_data['Adj Close'].pct_change()
    historical_data['PnL'] = historical_data['Signal'].shift(1) * historical_data['Return']
    historical_data['Cumulative_PnL'] = (1 + historical_data['PnL']).cumprod()

    # Calcul du Sharpe Ratio
    sharpe_ratio = historical_data['PnL'].mean() / historical_data['PnL'].std() * np.sqrt(252)

    # Calcul du drawdown
    historical_data['Peak'] = historical_data['Cumulative_PnL'].cummax()
    historical_data['Drawdown'] = (historical_data['Peak'] - historical_data['Cumulative_PnL']) / historical_data['Peak']
    max_drawdown = historical_data['Drawdown'].max()

    # Cration du carnet d'ordre
    orders = []
    for i in range(len(historical_data)):
        if not np.isnan(historical_data['Buy_Price'][i]):
            orders.append({'Date': historical_data.index[i], 'Type': 'Buy', 'Price': historical_data['Adj Close'][i], 'Units': 1})
        if not np.isnan(historical_data['Sell_Price'][i]):
            orders.append({'Date': historical_data.index[i], 'Type': 'Sell', 'Price': historical_data['Adj Close'][i], 'Units': 1})
    orders_df = pd.DataFrame(orders)

    # Graphique du prix de l'actif avec signaux d'achat et de vente
    plt.figure(figsize=(14, 7))
    plt.plot(historical_data['Adj Close'], label='Adj Close Price', color='grey')
    plt.scatter(historical_data.index, historical_data['Buy_Price'], label='Buy Signal', marker='^', color='green', alpha=1)
    plt.scatter(historical_data.index, historical_data['Sell_Price'], label='Sell Signal', marker='v', color='red', alpha=1)
    plt.title(f'Price of {symbol} with Buy/Sell Signals based on RSI')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Calcul des bnfices totaux
    buy_prices = historical_data['Buy_Price'].dropna().values
    sell_prices = historical_data['Sell_Price'].dropna().values
    num_trades = min(len(buy_prices), len(sell_prices))
    profits = sell_prices[:num_trades] - buy_prices[:num_trades]
    total_profit = profits.sum()
    plt.plot (np.array(historical_data['Cumulative_PnL']))
    plt.title('Cumulative PandL')
    plt.show()
    # Retourner les rsultats sous forme de dictionnaire
    return {
        'Sharpe_Ratio': sharpe_ratio,
        'Max_Drawdown': max_drawdown,
        'Total_Profit': total_profit,
        'Cumulative_PnL': historical_data['Cumulative_PnL'].iloc[-1], 
        'Order_Book': orders_df
    }

# Tester la stratgie sur le NASDAQ (IXIC) pour la priode entre 2020 et 2024
ftse_results = analyze_trading_strategy_RSI('^FTSE', '2000-01-01', '2013-06-01')
print("Results for FTSE:", ftse_results)

# Tester la stratgie sur le SPY (ETF S&P 500) pour la priode entre 2020 et 2024
spy_results = analyze_trading_strategy_RSI('SPY', '2000-01-01', '2013-06-01')
print("Results for SPY:", spy_results)
```

Bollinger bands strategy

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def bollinger_bands_strategy(symbol, start_date='2000-01-01', end_date='2013-06-01', window=20, num_std=2):
    # Tlchargement des donnes historiques
    data = yf.download(symbol, start=start_date, end=end_date)
    
    # Calcul des moyennes mobiles et des bandes de Bollinger
    data['SMA'] = data['Adj Close'].rolling(window=window).mean()
    data['Upper Band'] = data['SMA'] + num_std * data['Adj Close'].rolling(window=window).std()
    data['Lower Band'] = data['SMA'] - num_std * data['Adj Close'].rolling(window=window).std()
    
    # Gnration des signaux d'achat et de vente bass sur les bandes de Bollinger
    data['Position'] = np.where(data['Adj Close'] < data['Lower Band'], 1, np.where(data['Adj Close'] > data['Upper Band'], -1, 0))
    data['Position'] = data['Position'].diff().fillna(0)
    data['Buy_Signal_Price'] = np.where((data['Position'] == 1), data['Adj Close'], np.nan)
    data['Sell_Signal_Price'] = np.where((data['Position'] == -1), data['Adj Close'], np.nan)

    # Calcul du P&L quotidien et cumulatif
    data['Return'] = data['Adj Close'].pct_change()
    data['PnL'] = data['Position'].shift(1) * data['Return']
    data['Cumulative PnL'] = (1 + data['PnL']).cumprod()

    # Calcul du Sharpe Ratio
    sharpe_ratio = data['PnL'].mean() / data['PnL'].std() * np.sqrt(252)

    # Calcul du drawdown
    data['Peak'] = data['Cumulative PnL'].cummax()
    data['Drawdown'] = (data['Peak'] - data['Cumulative PnL']) / data['Peak']
    max_drawdown = data['Drawdown'].max()

    # Graphique du prix de l'actif avec signaux d'achat et de vente
    plt.figure(figsize=(14, 7))
    plt.plot(data['Adj Close'], label='Adj Close Price', color='blue')
    plt.plot(data['SMA'], label='Simple Moving Average', color='black')
    plt.plot(data['Upper Band'], label='Upper Bollinger Band', color='red')
    plt.plot(data['Lower Band'], label='Lower Bollinger Band', color='green')
    plt.scatter(data.index, data['Buy_Signal_Price'], label='Buy Signal', marker='^', color='green', alpha=1,s=100)
    plt.scatter(data.index, data['Sell_Signal_Price'], label='Sell Signal', marker='v', color='red', alpha=1,s=100)
    plt.title(f'{symbol} Price with Buy/Sell Signals based on Bollinger Bands')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Cration du carnet d'ordre
    orders = []
    for i in range(len(data)):
        if not np.isnan(data['Buy_Signal_Price'][i]):
            orders.append({'Date': data.index[i], 'Type': 'Buy', 'Price': data['Adj Close'][i], 'Units': 1})
        if not np.isnan(data['Sell_Signal_Price'][i]):
            orders.append({'Date': data.index[i], 'Type': 'Sell', 'Price': data['Adj Close'][i], 'Units': 1})
    orders_df = pd.DataFrame(orders)

    # Calcul des bnfices totaux
    buy_prices = data['Buy_Signal_Price'].dropna().values
    sell_prices = data['Sell_Signal_Price'].dropna().values
    num_trades = min(len(buy_prices), len(sell_prices))
    profits = sell_prices[:num_trades] - buy_prices[:num_trades]
    total_profit = profits.sum()
    plt.plot (np.array(data['Cumulative PnL']))
    plt.title('Cumulative PandL')
    # Retourner les rsultats sous forme de dictionnaire
    return {
        'Total Profit': total_profit,
        'Sharpe Ratio': sharpe_ratio,
        'Max Drawdown': max_drawdown,
        'Cumulative P&L': data['Cumulative PnL'].iloc[-1], 
        'Order Book': orders_df
    }

spy = bollinger_bands_strategy('SPY', start_date='2000-01-01', end_date='2013-06-01', window=20, num_std=2)
print(spy)

fse = bollinger_bands_strategy('^FTSE', start_date='2000-01-01', end_date='2013-06-01', window=20, num_std=2)
print(fse)
```
